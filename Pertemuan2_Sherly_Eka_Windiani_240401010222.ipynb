## 3. NumPy: Komputasi Numerik

### 3.1 Membuat NumPy Array

```python
import numpy as np
# Dari list Python
a1D = np.array([1, 2, 3, 4, 5]) # 1D
a2D = np.array([[1,2,3],[4,5,6]]) # 2D

# Array dengan nilai inisial
zeros = np.zeros((3, 4)) # nol semua
ones = np.ones((2, 3)) # satu semua
eye = np.eye(4) # matriks identitas
full = np.full((2, 2), 7) # nilai konstan

# Sequence
arange = np.arange(0, 20, 2) # 0,2,4,...18
linsp = np.linspace(0, 1, 11) # 11 titik merata

# Array acak (reproducible)
rng = np.random.default_rng(seed=42)
rand_int = rng.integers(0, 100, (3, 4))
rand_flt = rng.normal(loc=0, scale=1, size=(100,))
```

### 3.2 Operasi Vektorisasi dan Fungsi Statistik

```python
# Operasi aritmetika vektorisasi
a = np.array([1, 2, 3, 4, 5])
b = np.array([10, 20, 30, 40, 50])
print(a + b) # → [11 22 33 44 55]
print(a * b) # → [10 40 90 160 250]
print(a ** 2) # → [1 4 9 16 25]

# Universal functions (ufuncs)
np.sqrt(a) # akar kuadrat tiap elemen
np.exp(a) # eksponensial (e^x)
np.log(b) # logaritma natural

# Fungsi statistik
data = rng.normal(0, 1, 1000)
print(f'Mean: {np.mean(data):.4f}')
print(f'Std: {np.std(data):.4f}')
print(f'Min: {np.min(data):.4f}')
print(f'Max: {np.max(data):.4f}')
print(f'Q1: {np.percentile(data,25):.4f}')
print(f'Q3: {np.percentile(data,75):.4f}')
```

### 4.5 Groupby dan Agregasi

```python
# Python — Groupby & Agregasi

# Pola dasar: groupby → aggregation
df.groupby('Pclass')['Survived'].mean()

# Agregasi multiple kolom & fungsi
hasil = df.groupby('Pclass').agg(
total = ('PassengerId', 'count'),
selamat = ('Survived', 'sum'),
pct_selamat= ('Survived', 'mean'),
rata_usia = ('Age', 'mean'),
max_fare = ('Fare', 'max')
)
hasil = hasil.round(3)
print(hasil)

# GroupBy multi-kolom
df.groupby(['Pclass','Sex'])['Survived']\
.mean().unstack().round(3)
```

## Q2 Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%) per kolom?

```python
# Menghitung jumlah missing value per kolom
missing = df.isnull().sum()

# Menghitung persentase missing value per kolom
persentase = (missing / len(df)) * 100

print("Jumlah Missing Value")
print(missing)

print("\nPersentase Missing Value")
print(persentase)
```

Banyaknya data yang hilang pada kolom Cabin dan Age menunjukkan bahwa kualitas data pada kedua kolom tersebut perlu diperhatikan karena dapat memengaruhi hasil analisis yang menggunakan informasi kabin dan usia penumpang.

## Q3 Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas.

```python
persen_selamat = df["Survived"].mean() * 100

print(f"Persentase penumpang yang selamat: {persen_selamat:.2f}%")
```

## Q4 Filter penumpang wanita dari kelas 1 (Pclass=1). Berapa jumlahnya dan berapa rata-rata usianya?

```python
# Filter wanita kelas 1
wanita_kelas1 = df[
    (df["Sex"] == "female") &
    (df["Pclass"] == 1)
]

# Jumlah wanita kelas 1
jumlah_wanita = len(wanita_kelas1)

# Rata-rata usia
rata_usia = wanita_kelas1["Age"].mean()

print("Jumlah wanita kelas 1:", jumlah_wanita)
print("Rata-rata usia:", round(rata_usia, 2))
```

### Interpretasi Q4

**Terdapat 94 penumpang wanita yang berada pada kelas 1 dengan rata-rata usia 34,61 tahun**.

Hasil ini menunjukkan bahwa kelompok wanita kelas 1 didominasi oleh penumpang usia dewasa. Informasi ini dapat digunakan untuk analisis lebih lanjut mengenai hubungan antara usia, kelas penumpang, dan tingkat keselamatan.

## Q5 Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?

```python
# Mengelompokkan data berdasarkan kelas penumpang (Pclass)
# lalu menghitung rata-rata nilai Survived pada setiap kelas
keselamatan = df.groupby("Pclass")["Survived"].mean() * 100

print("Tingkat keselamatan berdasarkan kelas penumpang (%):")
print(keselamatan.round(2))
```

### Interpretasi Q5

**Tingkat keselamatan tertinggi terdapat pada penumpang kelas 1, yaitu sebesar 62,96%,** diikuti kelas 2 sebesar 47,28%, dan kelas 3 sebesar 24,24%.

Perbedaan ini menunjukkan bahwa penumpang dari kelas yang lebih tinggi memiliki peluang selamat yang lebih besar dibandingkan penumpang kelas bawah. Hal ini mengindikasikan bahwa kelas penumpang kemungkinan memiliki pengaruh terhadap tingkat keselamatan pada tragedi Titanic.
